```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X = df.drop("Disease", axis =1)
y = df['Disease']
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
```

```python
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 40)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
models = {
    "SVC" : SVC(kernel = 'linear'),
    "RandomForest": RandomForestClassifier(n_estimators = 100, random_state = 40),
    "GradientBoosting": GradientBoostingClassifier(n_estimators=100, random_state = 40),
    "MultinomialNB": MultinomialNB()
}  
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model_name}accuracy: {accuracy}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

# Assuming you have already prepared your X_train, y_train, X_test, y_test

# Initialize the Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')

# Fit the model
rf.fit(X_train, y_train)
```

```python
import pickle
with open("disease_prediction_model.pkl", "wb") as file:
    pickle.dump(rf, file)  # Save the model

print("Model saved as disease_prediction_model.pkl")
```

```python
y_pred = rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Random Forest Model Accuracy: {accuracy:.2f}")

diseases_list = {encoded_label: disease for encoded_label, disease in enumerate(le.classes_)}

all_symptoms = [symptom.strip() for symptom in X.columns]
symptoms_dict = {symptom: idx for idx, symptom in enumerate(all_symptoms)}
```

```python
# Save symptoms dictionary
with open("symptoms_dict.pkl", "wb") as file:
    pickle.dump(symptoms_dict, file)

print("Symptoms dictionary saved as symptoms_dict.pkl")
```

```python
# Save diseases list dictionary
with open("diseases_list.pkl", "wb") as file:
    pickle.dump(diseases_list, file)

print("Diseases list saved as diseases_list.pkl")
```

```python
def predict_diseases_with_probabilities_rf(patient_symptoms):
    
    patient_symptoms = [symptom.lower().replace(" ", "_") for symptom in patient_symptoms]
    input_vector = np.zeros(len(symptoms_dict))
    for symptom in patient_symptoms:
        if symptom in symptoms_dict:
            input_vector[symptoms_dict[symptom]] = 1
        else:
            print(f"Warning: Symptom '{symptom}' not found in training data.")
    
    # Get probabilities from the Random Forest model
    predicted_probs = rf.predict_proba([input_vector])[0]
    
    # Map probabilities to disease names
    predicted_diseases = {
        diseases_list[label]: prob 
        for label, prob in enumerate(predicted_probs) if prob > 0.01
    }
    
    # Sort diseases by probability in descending order
    sorted_diseases = sorted(predicted_diseases.items(), key=lambda x: x[1], reverse=True)[:5]
    
    return sorted_diseases
```

```python
# Example input symptoms
input_symptoms = ['stomach pain',  'acidity', 'vomitting','cough']  # Example inputs
predicted_diseases = predict_diseases_with_probabilities_rf(input_symptoms)

# Display predictions
print("Predicted Diseases with Probabilities:")
for disease, prob in predicted_diseases:
    print(f"{disease}: {prob * 100:.2f}%")
```

```python
df_description = pd.read_csv("symptom_Description.csv")
df_description.head()
```

```python
df_precautions = pd.read_csv("symptom_precaution.csv")
df_precautions.head()
```

```python
diseases_2_top = [disease[0] for disease in predicted_diseases[:2]]

def disease_description(input_disease):
    description_dict = {}
    for disease in input_disease:
        description = df_description[df_description["Disease"]== disease]["Description"].tolist()
        description_dict[disease] = description[0] if description else "Description not found"
    return description_dict
```

```python
from sklearn.impute import SimpleImputer
# Handle missing values
imputer = SimpleImputer(strategy="most_frequent")
df_precautions[["Precaution_1", "Precaution_2", "Precaution_3","Precaution_4"]] = imputer.fit_transform(df_precautions[["Precaution_1", "Precaution_2", "Precaution_3", "Precaution_4"]])

def get_precautions(disease_list):
    precautions_dict = {}
    for disease in disease_list:
        precautions = df_precautions[df_precautions["Disease"] == disease].values.flatten().tolist()[1:]
        precautions_dict[disease] = [p for p in precautions if pd.notna(p)]
    return precautions_dict
```

```python
descriptions = disease_description(diseases_2_top)
precautions = get_precautions(diseases_2_top)

print("Disease Predictions and Recommendations:\n")
for disease in diseases_2_top:
    print(f"**Disease:** {disease}")
    print(f"**Description:** {descriptions[disease]}\n")
    print(f"**Precautions:**")
    for i, precaution in enumerate(precautions[disease], 1):
        print(f"{i}. {precaution}")
    print("\n" + "-"*50 + "\n") 
```
