```python
import pandas as pd
```

```python
df = pd.read_csv("datasetdoc.csv")
```

```python
df.head()
```

```python
all_symptoms = df.iloc[:, 1:].stack().unique()
all_symptoms = [symptom for symptom in all_symptoms if pd.notna(symptom)]  # Remove NaN

print(all_symptoms)
```

```python
for symptom in all_symptoms:
    df[symptom] = 0
for index, row in df.iterrows():
    for col in df.columns[1:18]:  # Iterate over symptom columns
        if pd.notna(row[col]):
            df.loc[index, row[col]] = 1
```
